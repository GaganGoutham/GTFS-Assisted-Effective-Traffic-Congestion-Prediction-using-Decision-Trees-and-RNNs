```python
import pandas as pd
import numpy as np
```

```python
peak = pd.read_csv('Slot_9-9.20.csv')
non_peak = pd.read_csv('Slot_7-7.20.csv')
```

```python
peak['SRI']=0.0
peak
```

```python
for i in range(0,20):
  for j in range(0,20):
    if((peak['stop_id_modified_from'][i]==non_peak['stop_id_modified_from'][j]) & (peak['stop_id_modified_to'][i]==non_peak['stop_id_modified_to'][j])):
      SRI=(1-(peak['avera'][i])/non_peak['average_time'][j])*10
      peak['SRI'][i]=SRI

peak["SRI"] = pd.to_numeric(peak["SRI"], downcast="float")     
```

```python
peak.to_csv('SRI1.csv')
```
