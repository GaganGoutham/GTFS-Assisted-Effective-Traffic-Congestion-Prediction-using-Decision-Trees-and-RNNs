```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("sri1.csv")
```

```python
for i in range(0,len(dataset['SRI'])):
  if(dataset['SRI'][i]==0):
    dataset['SRI'][i]=(1-(dataset['speed'][i])/23)*10
    
```

```python
dataset.tail()
```

```python
dataset['Degree_of_congestion']=''
```

```python
for i in range(0,len(dataset['SRI'])):
  if(float(dataset['SRI'][i])<=0):
    dataset['Degree_of_congestion'][i]='Very smooth'
  elif((float(dataset['SRI'][i])>0) & (float(dataset['SRI'][i])<=2.5)):
    dataset['Degree_of_congestion'][i]='Smooth'  
  elif((float(dataset['SRI'][i])>2.5) & (float(dataset['SRI'][i])<=5)):
    dataset['Degree_of_congestion'][i]='Mild congestion'
  else:
    dataset['Degree_of_congestion'][i]='Heavy congestion'          
```

```python
if(float(dataset['SRI'][0])<=0):
    dataset['Degree_of_congestion']='Very smooth'
```

```python
stop_times = pd.read_csv('18.20mins_slot.csv')
```

```python
dataset['arrival_time']=''
```

```python
for i in range(11898,len(dataset['SRI'])):
  for j in range(0,len(stop_times['stop_id'])):
    if((dataset['stop_id_from'][i]==stop_times['stop_id'][j]) & (dataset['trip_id'][i]==stop_times['trip_id'][j])):
      dataset['arrival_time'][i] = stop_times['arrival_time'][j] 
```
