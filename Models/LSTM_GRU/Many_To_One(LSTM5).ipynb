```python
from numpy import array
from keras.preprocessing.text import one_hot
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers.core import Activation, Dropout, Dense
from keras.layers import Flatten, LSTM
from keras.layers import GlobalMaxPooling1D
from keras.models import Model
from keras.layers.embeddings import Embedding
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import Tokenizer
from keras.layers import Input
from keras.layers.merge import Concatenate
from keras.layers import Bidirectional

import pandas as pd
import numpy as np
import re

import matplotlib.pyplot as plt
```

# Many To One - Simple (**17.82%**)

```python
X = array(X).reshape(13298,5,4)
```

```python
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(5, 4)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse',metrics=["accuracy"])
```

```python
trial = model.fit(X, y, epochs=1000, validation_split=0.2, verbose=1)
```

# Many To One - Stacked (**17.81%**)

```python
model = Sequential()
model.add(LSTM(200, activation='relu', return_sequences=True, input_shape=(5, 4)))
model.add(LSTM(100, activation='relu', return_sequences=True))
model.add(LSTM(50, activation='relu', return_sequences=True))
model.add(LSTM(25, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse',metrics=["accuracy"])
```

```python
trial1 = model.fit(X, y, epochs=100, validation_split=0.2, verbose=1)
```

# Many To One - Bidirectional (**17.82%**)

```python
from keras.layers import Bidirectional

model1 = Sequential()
model1.add(Bidirectional(LSTM(50, activation='relu'), input_shape=(5, 4)))
model1.add(Dense(1))
model1.compile(optimizer='adam', loss='mse',metrics=["accuracy"])
```

```python
trial2 = model1.fit(X, y, epochs=100, validation_split=0.2, verbose=1)
```
