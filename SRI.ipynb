```python
import pandas as pd
import numpy as np
```

```python
peak = pd.read_csv('slot_9_00-9_20_trips.csv')
non_peak = pd.read_excel('slot_07_00-07_20_trips.xlsx')
```

```python
peak['SRI']=0
peak
```

```python
for i in range(0,len(peak['stop_id_from'])):
  for j in range(0,len(non_peak['stop_id_modified_from'])):
    if((peak['stop_id_from'][i]==non_peak['stop_id_modified_from'][j]) & (peak['stop_id_to'][i]==non_peak['stop_id_modified_to'][j])):
      SRI=(1-(peak['speed'][i])/non_peak['speed'][j])*10
      peak['SRI'][i]=SRI

peak["SRI"] = pd.to_numeric(peak["SRI"], downcast="float")      
```

```python
peak.to_csv('sri1.csv')
```

```python
for i in range(0,10):
  for j in range(0,11):
    if((peak['stop_id_from'][i]==non_peak['stop_id_modified_from'][j]) & (peak['stop_id_to'][i]==non_peak['stop_id_modified_to'][j])):
      SRI=(1-(peak['speed'][i])/23)*10
      peak['SRI'][i]=SRI

peak["SRI"] = pd.to_numeric(peak["SRI"], downcast="float")      
```
